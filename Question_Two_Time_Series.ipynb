```python
#Selecting the column and renaming the 'close' column
df_btc_selected= btc_df[['date','close']]
df_btc_selected=df_btc_selected.copy()
df_btc_selected.rename(columns={
    'close':'btc_price'
},inplace=True)
df_btc_selected
```

```python
#merging gold price and bitcoin price
df_gold_selected=df_gold_selected.copy()
df_btc_selected=df_btc_selected.copy()
df_gold_selected['date']=df_gold_selected['date'].astype(str)
df_btc_selected['date']=df_btc_selected['date'].astype(str)
final_df=pd.merge(df_gold_selected,df_btc_selected,how='left',on=['date'])
final_df
```

```python
#dropping null values
final_df.dropna(inplace=True)
final_df
```

```python
final_df['date'] = pd.to_datetime(final_df['date'])
final_df
```

```python
#xt is gold price and yt is bitcoin(BTC) price
xt=final_df['gold_price'].to_numpy()
yt=final_df['btc_price'].to_numpy()
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#Calculating alpha and beta
beta=((np.multiply(yt,xt)).mean()-np.mean(xt)*np.mean(yt))/((np.multiply(xt, xt)).mean()-np.mean(xt)*np.mean(xt))
alpha=np.mean(yt)-beta*np.mean(xt)
```

```python
print("alpha =",alpha, "\nbeta =",beta)
```

```python
#Calculating y hat.
yt_hat=alpha + np.multiply(beta,xt)
L = np.sum(np.multiply(yt-yt_hat,yt-yt_hat))
L
```

```python
#calculating mean square error
def cal_cost(alpha,beta,xt,yt):
    n = len(yt)
    predictions= alpha + np.multiply(beta,xt)
    cost=(1/2*n) * np.sum(np.square(predictions-yt))
    return cost
    
```

```python
cal_cost(alpha,beta,xt,yt)
```

```python
#Gradient Descent
iterations = 1000
alpha = -11
n = len(yt)

loss_history = np.zeros((iterations,1))
alpha_history = np.zeros((iterations,1))

for i in range(iterations):
    
    prediction = alpha + np.multiply(beta,xt)
    residuals = yt-prediction
    residuals_sum = sum(residuals)
    gradient = -(1/n)*0.1*residuals_sum
    
    alpha = alpha - gradient
    
    loss_history[i] = cal_cost(alpha,beta,xt,yt)
    alpha_history[i] = alpha
    
```

```python
plt.plot(alpha_history,loss_history,'go')
```
